```
import tensorflow as tf
from tensorflow import keras


from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

# 1 Get the Data (sentences)

```
sentences = [
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think my dog is amazing?'
]
```

# 2 Tokenizer via `from tensorflow.keras.preprocessing.text import Tokenizer`

## Fit the sentences to create tokens

```
tokenizer = Tokenizer(num_words = 100, oov_token="<OOV>") # handling out of vocabulary words
tokenizer.fit_on_texts(sentences)
```

## Get word indices

```
word_index = tokenizer.word_index
word_index
```

## Turn each sentence to indexed sequences

```
# list of lists - each list is an indexed sentence
sequences = tokenizer.texts_to_sequences(sentences) 
sequences
```

# 3 Padding and Truncation via `from tensorflow.keras.preprocessing.sequence import pad_sequences`

## Different Paddings

```
# post padding

padded = pad_sequences(sequences, padding="post")
print("\nWord Index = " , word_index)
print("\nSequences = " , sequences)
print("\nPadded Sequences:")
print(padded)
```

```
# default is pre padding

padded = pad_sequences(sequences)
print("\nWord Index = " , word_index)
print("\nSequences = " , sequences)
print("\nPadded Sequences:")
print(padded)
```

## Different Truncations

```
# post padding
# and truncate at the start

padded = pad_sequences(sequences, padding="post", maxlen = 5)
print("\nWord Index = " , word_index)
print("\nSequences = " , sequences)
print("\nPadded Sequences:")
print(padded)
```

```
# post padding
# and truncate at the end

padded = pad_sequences(sequences, padding="post", truncating = "post", maxlen = 5)
print("\nWord Index = " , word_index)
print("\nSequences = " , sequences)
print("\nPadded Sequences:")
print(padded)
```

# 4 Trying on Test Data

```
# Try with words that the tokenizer wasn't fit to
test_data = [
    'i really love my dog',
    'my dog loves my manatee'
]

test_seq = tokenizer.texts_to_sequences(test_data)
print("\nTest Sequence = ", test_seq)

padded = pad_sequences(test_seq, maxlen=10)
print("\nPadded Test Sequence: ")
print(padded)
```
